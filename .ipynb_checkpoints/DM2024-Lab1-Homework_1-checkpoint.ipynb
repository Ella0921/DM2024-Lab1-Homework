```python
# Exercise 11

# Answer here

# Threshold: Only include terms that appear in at least 3 documents
term_threshold = 3
document_threshold = 3

from sklearn.feature_extraction.text import TfidfTransformer

# Transform the term-document matrix to TF-IDF
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(X_counts).toarray()

# Sum across rows (documents) and columns (terms) to identify those with enough occurrences
non_zero_rows = np.where((tfidf_matrix > 0).sum(axis=1) >= document_threshold)[0]
non_zero_cols = np.where((tfidf_matrix > 0).sum(axis=0) >= term_threshold)[0]

# Subset the matrix after applying the threshold
filtered_matrix = tfidf_matrix[non_zero_rows][:, non_zero_cols]

# Get the corresponding document and term labels after filtering
filtered_plot_x = ["term_" + str(i) for i in np.array(count_vect.get_feature_names_out())[non_zero_cols]]
filtered_plot_y = ["doc_" + str(i) for i in np.array(X.index)[non_zero_rows]]

# Randomly sample 50 documents and 50 terms
sample_size = 50
sampled_rows = np.random.choice(filtered_matrix.shape[0], size=min(sample_size, filtered_matrix.shape[0]), replace=False)
sampled_cols = np.random.choice(filtered_matrix.shape[1], size=min(sample_size, filtered_matrix.shape[1]), replace=False)

# Subset the filtered matrix with the sampled rows and columns
sampled_matrix = filtered_matrix[sampled_rows][:, sampled_cols]

# Get the corresponding labels for the sampled rows and columns
sampled_plot_x = [filtered_plot_x[i] for i in sampled_cols]
sampled_plot_y = [filtered_plot_y[i] for i in sampled_rows]

# Create DataFrame for the sampled matrix and plot it
df_sampled_todraw = pd.DataFrame(sampled_matrix, columns=sampled_plot_x, index=sampled_plot_y)

# Plot the sampled matrix
plt.subplots(figsize=(12, 10))
sns.heatmap(df_sampled_todraw, cmap="PuRd", vmin=0, vmax=1, annot=False)
plt.title('Sampled Term-Document Matrix (50 Documents and 50 Terms)')
plt.show()
```

## 3. Attempt the tasks on the new dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from wordcloud import WordCloud
```

```python
# Load data
df = pd.read_csv("hf://datasets/Senem/Nostalgic_Sentiment_Analysis_of_YouTube_Comments_Data/Nostalgic_Sentiment_Analysis_of_YouTube_Comments_Data.csv")
df = df.dropna()
df.head()
```

```python
# Distribution of comment lengths for each sentiment category
df['comment_length'] = df['comment'].apply(len)
plt.figure(figsize=(10, 6))
sns.boxplot(x='sentiment', y='comment_length', data=df)
plt.title('Distribution of comment lengths by sentiment category')
plt.xlabel('Sentiment')
plt.ylabel('Comment length')
plt.show()
```

```python
# Combine all comments into a string
all_comments = " ".join(comment for comment in df['comment'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color="white").generate(all_comments)

# Show word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Common words in comments word cloud")
plt.show()
```

```python
# Generate features using word frequency
count_vectorizer = CountVectorizer(max_features=1000)
X_count = count_vectorizer.fit_transform(df['comment'])

# Generate features using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
X_tfidf = tfidf_vectorizer.fit_transform(df['comment'])

# Label
y = df['sentiment']

# Divide the training set and test set
X_count_train, X_count_test, y_train, y_test = train_test_split(X_count, y, test_size=0.2, random_state=42)
X_tfidf_train, X_tfidf_test = train_test_split(X_tfidf, test_size=0.2, random_state=42)
```

```python
# Multinomial Naive Bayes using word frequency features
nb_count = MultinomialNB()
nb_count.fit(X_count_train, y_train)
y_count_pred = nb_count.predict(X_count_test)

# Evaluating the word frequency feature model
count_accuracy = accuracy_score(y_test, y_count_pred)
print("MultinomialNB accuracy based on word frequency: ", count_accuracy)
print("Confusion matrix:\n", confusion_matrix(y_test, y_count_pred))
print("Classification report:\n", classification_report(y_test, y_count_pred))
```

```python
# Convert sparse matrix to dense matrix
X_tfidf_train_dense = X_tfidf_train.toarray()
X_tfidf_test_dense = X_tfidf_test.toarray()

# Gaussian Naive Bayes using TF-IDF features
nb_tfidf = GaussianNB()
nb_tfidf.fit(X_tfidf_train_dense, y_train)
y_tfidf_pred = nb_tfidf.predict(X_tfidf_test_dense)

# Evaluating the TF-IDF feature model
tfidf_accuracy = accuracy_score(y_test, y_tfidf_pred)
print("\nGaussianNB accuracy based on TF-IDF: ", tfidf_accuracy)
print("Confusion matrix:\n", confusion_matrix(y_test, y_tfidf_pred))
print("Classification report:\n", classification_report(y_test, y_tfidf_pred))
```

# Result Analysis

On the word frequency feature, Multinomial Naive Bayes performs better because it is suitable for discrete data; on the TF-IDF feature, Gaussian Naive Bayes is suitable for continuous data features.
