```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from wordcloud import WordCloud
```

```python
# Load data
df = pd.read_csv("hf://datasets/Senem/Nostalgic_Sentiment_Analysis_of_YouTube_Comments_Data/Nostalgic_Sentiment_Analysis_of_YouTube_Comments_Data.csv")
df = df.dropna()
df.head()
```

```python
# Distribution of comment lengths for each sentiment category
df['comment_length'] = df['comment'].apply(len)
plt.figure(figsize=(10, 6))
sns.boxplot(x='sentiment', y='comment_length', data=df)
plt.title('Distribution of comment lengths by sentiment category')
plt.xlabel('Sentiment')
plt.ylabel('Comment length')
plt.show()
```

```python
# Combine all comments into a string
all_comments = " ".join(comment for comment in df['comment'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color="white").generate(all_comments)

# Show word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Common words in comments word cloud")
plt.show()
```

```python
# Generate features using word frequency
count_vectorizer = CountVectorizer(max_features=1000)
X_count = count_vectorizer.fit_transform(df['comment'])

# Generate features using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
X_tfidf = tfidf_vectorizer.fit_transform(df['comment'])

# Label
y = df['sentiment']

# Divide the training set and test set
X_count_train, X_count_test, y_train, y_test = train_test_split(X_count, y, test_size=0.2, random_state=42)
X_tfidf_train, X_tfidf_test = train_test_split(X_tfidf, test_size=0.2, random_state=42)
```

```python
# Multinomial Naive Bayes using word frequency features
nb_count = MultinomialNB()
nb_count.fit(X_count_train, y_train)
y_count_pred = nb_count.predict(X_count_test)

# Evaluating the word frequency feature model
count_accuracy = accuracy_score(y_test, y_count_pred)
print("MultinomialNB accuracy based on word frequency: ", count_accuracy)
print("Confusion matrix:\n", confusion_matrix(y_test, y_count_pred))
print("Classification report:\n", classification_report(y_test, y_count_pred))
```

```python
# Convert sparse matrix to dense matrix
X_tfidf_train_dense = X_tfidf_train.toarray()
X_tfidf_test_dense = X_tfidf_test.toarray()

# Gaussian Naive Bayes using TF-IDF features
nb_tfidf = GaussianNB()
nb_tfidf.fit(X_tfidf_train_dense, y_train)
y_tfidf_pred = nb_tfidf.predict(X_tfidf_test_dense)

# Evaluating the TF-IDF feature model
tfidf_accuracy = accuracy_score(y_test, y_tfidf_pred)
print("\nGaussianNB accuracy based on TF-IDF: ", tfidf_accuracy)
print("Confusion matrix:\n", confusion_matrix(y_test, y_tfidf_pred))
print("Classification report:\n", classification_report(y_test, y_tfidf_pred))
```

### Result Analysis

On the word frequency feature, Multinomial Naive Bayes performs better because it is suitable for discrete data; on the TF-IDF feature, Gaussian Naive Bayes is suitable for continuous data features.

## 4. Solutions to efficiently handle different dataset



### Dataset 1: Original Dataset from Lab
In the lab, we quickly implemented the preprocessing steps for the original dataset, and there were some areas where the process was not ideal:

1. **Generalized Missing Value Imputation**:
   - **Inefficiency**: We used a generalized strategy to impute missing values, either using mean or median for numerical variables. This approach might not capture the true nature of the dataset when certain features have skewed distributions or outliers.
   - **Improvement**: A better solution would be to analyze the distribution of each feature before deciding on an imputation method. We could use **median imputation for skewed features** and consider advanced techniques such as **K-Nearest Neighbors (KNN) imputation** to incorporate relationships between different features for more accurate estimates.

2. **Uniform Scaling Strategy**:
   - **Inefficiency**: We applied the same scaling method (standardization) to all numerical features without considering their individual properties. This caused certain features to lose valuable information, particularly those with a limited range or those heavily influenced by outliers.
   - **Improvement**: Instead of a uniform approach, we could utilize **Min-Max scaling** for features with bounded ranges and **robust scaling** for features with extreme outliers, thus preserving important variability while ensuring comparability.

3. **One-Hot Encoding for All Categorical Features**:
   - **Inefficiency**: We used **One-Hot Encoding** for all categorical features, which led to a significant increase in dimensionality, especially for features with high cardinality.
   - **Improvement**: To reduce dimensionality and increase efficiency, we could use **Target Encoding** or **Frequency Encoding** for high-cardinality features, ensuring we capture meaningful information while managing dataset size.

4. **No Handling of Feature Correlations**:
   - **Inefficiency**: We did not perform feature correlation analysis, which could lead to multicollinearity and reduce the interpretability of our model.
   - **Improvement**: We could calculate **Pearson’s correlation** between numerical features and **remove or combine highly correlated features**. This would help prevent redundant information, making the model simpler and more efficient.

### Dataset 2: New Dataset Introduced in Questions 2 and 3
For the new dataset introduced in questions 2 and 3, the quick implementation of preprocessing steps also resulted in some inefficiencies:

1. **Lack of Custom Handling for Text Features**:
   - **Inefficiency**: The new dataset contained text features that were not adequately preprocessed. We applied basic tokenization but did not consider other preprocessing steps like removing stop words or stemming.
   - **Improvement**: We should apply **text cleaning** to remove punctuation, stop words, and perform **stemming or lemmatization** to normalize words. Additionally, using **TF-IDF vectorization** instead of simple token counts would help capture the importance of words while reducing dimensionality.

2. **Mixed Handling of Different Data Types**:
   - **Inefficiency**: The new dataset contained both numerical and categorical features, but our preprocessing did not differentiate adequately between them. For example, we scaled categorical numerical features that should have been treated differently.
   - **Improvement**: Instead of scaling all numerical features uniformly, we need to identify and separate **ordinal categorical features** (which need **label encoding**) from continuous numerical features (which need **scaling**). This ensures that categorical features are properly interpreted by the model.

3. **Ignoring Imbalance in the Target Variable**:
   - **Inefficiency**: We did not check for class imbalance in the target variable, which could lead to poor model performance, particularly on the minority class.
   - **Improvement**: We should perform an **imbalance check** and, if necessary, use techniques such as **SMOTE (Synthetic Minority Over-sampling Technique)** or **undersampling** to address imbalance and ensure the model is not biased towards the majority class.

4. **Feature Selection Based on Statistical Significance**:
   - **Inefficiency**: We did not perform any feature selection, which led to including irrelevant or weakly correlated features. This increased the model complexity and might negatively affect its generalization.
   - **Improvement**: We could use feature selection techniques like **Recursive Feature Elimination (RFE)** or **mutual information analysis** to select only the most relevant features. This would improve model training efficiency and potentially lead to better performance.
